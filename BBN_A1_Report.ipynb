```python
import matplotlib.pyplot as plt
import numpy as np
from math import*

eta = np.geomspace(10**(-12.1), 10**(-7), 1000)
eta_1 = [x * (10**(10)) for x in eta]

T_c = 0.03 / ((1 + np.log(eta_1) / 7)**3)
T_n = 0.25 / (1 + np.log(eta_1) / 35)

plt.figure(figsize=(8, 6), dpi=300)
plt.semilogx(eta, T_c, label=r'$T_c$', color='red')
plt.semilogx(eta, T_n, label=r'$T_{nuclei}$', color='blue')
plt.axhline(y=0.02, color='darkblue', linestyle='--', label=r'$T_n$')
plt.axhline(y=1, color='olive', linestyle='dotted', label=r'$T_F$')
plt.annotate(r'$NSE+\beta$ Equilibrium', xy=(10**(-10.7), 1.005))

plt.xlabel(r'$\eta$')
plt.ylabel(r'$T/MeV$')
plt.grid(True)
plt.legend()
plt.title("BBN Temperature V/S n-p ratio")
plt.tight_layout()
plt.yscale('log')
plt.show()
```

* Baryonic Matter/Dark Matter dominated Universe $\rho=CR^{-3}$
* Universe dominated by Vacuum Energy $\rho=constant$

```python
import numpy as np
import matplotlib.pyplot as plt

c_r = 2.60 * 10**(-10)
c_b = 2.34 * 10**(-7)
c_d = 1.25 * 10**(-6)
c_v = 3.2 * 10**(-6)
r = np.geomspace(10**(-10), 10**(0), 10000)
c_m = c_b + c_d
float_number = c_r / c_b
exp_number = "{:e}".format(float_number)
f_2 = c_r / c_m
y2 = c_r * (f_2**(-4))
x2 = f_2
y = c_r * (float_number**(-4))
x = float_number

p1 = c_r * r**(-4)
p2 = c_b * r**(-3)
p3 = c_d * r**(-3)
p4 = c_v
p5 = c_m * r**(-3)

plt.figure(figsize=(8, 6), dpi=300)
plt.xscale("log")
plt.yscale("log")
#plt.plot(x, y, 'go')
plt.plot(r, p1, color="red", label=r'$\rho_{r} $ (radiation density)')
plt.plot(r, p5, color="blue", label=r'$\rho_{m} $ (matter density)')

plt.ylabel(r'$\rho( GeV/cm^{3})$', fontsize=16)
plt.xlabel(r'$R$', fontsize=16)
plt.plot(x2, y2, 'go')
plt.annotate(('(1.75e-04,2.76e+05)'), (x2, y2), fontsize=12)
plt.axhline(y=p4, linestyle="dotted", label=r'$\rho_{v}$ (vacuum energy)', color="black")
plt.plot(x3, y3, 'yo')
plt.annotate(('Photon Decoupling'), (x3+10**-3, y3), fontsize=10, color='orange')
plt.axvline(x=x2, linestyle="--", color="olive")
plt.annotate(('Radiation Dominated Era'), (10**(-9),10**(5)), fontsize=12, color='grey')
plt.annotate(('Radiation Dominated Era'), (10**(-9),10**(5)), fontsize=12, color='grey')

plt.annotate(('Matter Dominated Era'), (10**(-3),10**(15)), fontsize=12, color='grey')



plt.grid(True)
plt.legend(fontsize=16)
plt.tight_layout()
plt.show()
```

* One thing is clear in the above graph: matter dominance only occurred after the process of Big Bang Nucleosynthesis (BBN). This indicates that BBN must have occurred at a significant distance from the point where the two curves intersect. Additionally, the expected time frame for BBN to occur is approximately $t \sim 10^{-3}-10^{2}$ seconds, with a corresponding redshift of $z \sim 10^{9}$, which translates to a scale factor of $R \sim 10^{-9}$.

* Photon decoupling took place roughly $t \sim 10^{5}$ seconds after the Big Bang. During this period, electrons and protons in the universe combined to form neutral hydrogen atoms. However, the electrons did not immediately reach the ground state. Instead, they first entered a higher excited state and then, by releasing energy, transitioned to a lower energy state, emitting photons in the process. As these neutral hydrogen atoms were mostly transparent to various frequencies, the photons were able to travel large distances. They can still be detected today as the Cosmic Microwave Background (CMB), which actually occurred at a redshift of $z \sim 1100$, corresponding to a scale factor of $R \sim 10^{-3}$.
